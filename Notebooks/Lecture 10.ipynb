## Dynamic Models

* Application of agent-based modeling (ABM) to simulating *dynamics* within geographical systems has seen considerable increase over last decade

* Overall Idea?
    * ABMs allow disaggregation of *systems* into *individual components*
    * Have own characteristics and rule sets
    * Builds on ideas of complexity, emergence, agency, and simulation

## Geographical Systems

* Geographical *systems* characterized by continual change and evolution through time and space
* The impacts of *interactions*
    * Individual agents $\leftrightarrow$ agents (humans, cities or more abstract representations)
    * Individual agent $\leftrightarrow$ environment (physical, social, information etc)
    * Manifest at multiple spatial scales and over differing time-scales

## Static Dynamics

* Traditional models of complexity of geographical systems
    * Focused on static aggregations of populations, rational aggregate behavior and flows of information
    * Multiple regression, location-allocation, spatial interaction models, etc.
* Makes understanding dynamics difficult
    * Doesn't really treat 'process' directly
    * Computationally simple $\neq$ better

## From Individuals $\rightarrow$ System

* Bottom-up approaches to simulating human/natural systems
    * Specific focus on *individuals*
* One such approach is agent-based modeling (ABM)
    * Simulate individual actions of diverse agents
    * Measure resulting *system behavior* and outcomes over time
* Main distinction?

## Process Driven

> Facilitate the exploration of system processes at the level of their constituent elements

## Constituent Elements

* Essential to progression of ABM... development of **automata** approaches

> An automaton is a processing mechanism with characteristics that change over time based on its internal characteristics, rules and external input...

## Automata

* Automata process information
    * Based on... inputs from their surroundings
    * Updates to... their internal characteristics
    * According to... rules that govern their reaction to inputs
* Two classes of automata frameworks:
    Cellular automata (CA) and agent-based models (ABM)

## What Is an Agent?

* Well, kinda almost *anything* really
    * No universal 'model'... but some key features $\hookrightarrow$

## Features of Agents

1. Autonomy
2. Heterogeneity
3. Active
    * Pro-active/goal-directed
    * Reactive/Perceptive
    * Bounded Rationality
    * Interactive/Communicative
    * Mobility
4. Adaptive

### Autonomy

* Agents are autonomous units
    * Governed without influence of centralized control
    * Capable of processing \& exchanging info /w other agents
        * Make independent decisions
    * Free to interact with other agents\*
        * Does not (necessarily) affect their autonomy
        
<br/>
<div class="alert alert-info", style="font-size: 0.8em;">
\* At least over a limited range of situations
</div>

### Heterogeneity

* Agents permit development of autonomous individuals
    * An agent representing a human could have attributes such as `age`, `sex`, `job` etc.
* Groups of agents can also exist...
    * Are *spawned* from bottom-up...
    * Are *amalgamations* of similar autonomous *individuals*

### Active (exert independent influence)

* Pro-active/goal-directed
    * Having goals to achieve with respect to behaviors
* Reactive/Perceptive
    * Awareness or sense of surroundings
    * May have prior knowledge
        * 'mental map' of environment, awareness of other entities, obstacles
        * required destinations within their environment

### Active (cont'd)

* Bounded Rationality
    * Can be configured with 'bounded' rationality (through their heterogeneity)\*
    * Allows agents to make inductive, discrete, and adaptive choices that $\rightarrow$ achieving goals
        
<br/>
<div class="alert alert-info", style="font-size: 0.8em;">
\* This is quite different from 'classical' social science, which is often based on a rational-choice paradigm...
</div>

### Active (cont'd)

* Interactive/Communicative
    * Ability to communicate extensively
        e.g., agents can query other agents and/or the environment within a neighborhood
* Mobility
    * Can 'roam' space within model, provides vast range of interactions
    * However, agents can also be fixed if needed...

### Adaptive

* Agents *can be* adaptive
    * Complex Adaptive Systems (CAS)
    * Alter state depending on previous... like memory
        * e.g., learning fastest/safest routes
* Adapt at individual level
    * e.g. learning alters prob dist of rules
* Or population level
    * e.g. learning alters freq dist of agents competing for reproduction

## Object Orientation

* Object-oriented or object-orientation is a software engineering concept
    * Concepts are represented as "objects"
* Lots of areas where objects are used, including databases, modeling, programing, design, etc.

## In General Terms

* Paradigm based on concept of "objects"
    * May contain data, in the form of *fields*, often known as attributes
    * and code, in the form of procedures, often known as methods
*An object's procedures can access/modify own (self) data fields
* OOP are designed by making them out of objects that interact with one another

## Core OO Concepts

* Objects/classes
* Encapsulation
* Inheritance
* Composition
* Polymorphism

### Objects and classes

* Classes
    * Defs for data format and available procedures for given type (class) of object
    * May contain data & procedures (class *methods*)
        * i.e. classes contains data members & member functions
* Objects (instances of classes)
    * Sometimes correspond to *things* found in 'real world'
    * e.g., "shopping cart", "customer", "product"

### Important Terms

* Each *object* is said to be an instance of a particular *class*
    * e.g., object with its name field set to `"Mary"` might be an instance of class `Employee`
* `Procedures` in object-oriented programming are known as methods
* `Variables` are also known as fields, members, attributes, or properties

### Encapsulation

* Internal representation of object is hidden from view outside of the object's definition
    * Typically, only object's own methods can directly inspect or manipulate its fields
* Closely related to concept of 'Abstraction'
    * Hides away complexity of implementation to simplify interaction

### Inheritance

* An object or class is based on another object (prototypal inheritance) or class (class-based inheritance)
    * An inherited class is called a subclass of its parent class
* Allows classes to be arranged in a hierarchy that represents "is-a-type-of" relationships
    * e.g., class `Employee` might inherit from class `Person`

### Composition

* Objects can contain other objects in their instance variables
    * e.g., object in `Employee` class might contain (point to) object in `Address` class
    * Used to represent "has-a" relationships:
        * Every employee *has an* address, so every `Employee` object *has a* place to store an `Address` object

### Polymorphism

* Polymorphism: one name, many forms
* Multiple methods all with the same name, but slightly different functionality
    * Calling code can be agnostic as to whether an object belongs to a parent class or one of its descendants
* Polymorphism is most commonly used when dealing with inheritance

## A Quick Python Example

```python
class Animal:
    def __init__(self, name=''):
        self.name = name
    
    def talk(self):
        pass  # Don't do anything

class Cat(Animal):
    def talk(self):
        print("Meow!")

class Dog(Animal):
    def talk(self):
        print("Woof!")
```

```python
a = Animal()
a.talk()
```

```python
c = Cat("Susan")
c.talk()
```

```python
d = Dog("Lucas")
d.talk()
```

```python
class House:
    def __init__(self, pets):
        self.pets = pets
        
    def ring_doorbell(self):
        for pet in self.pets:
            pet.talk()
```

```python
house = House([c, a])
house.ring_doorbell()
```

## Rules, Behavior and Relationships

* Agents can possess rules that affect behavior and relationships with...
    * Other agents and/or their surrounding environment
* Rules are typically derived from published literature, expert knowledge, data analysis or numerical work
    * Are the foundation of an agent's behavior
* One rule-set can be applied to...
    * All agents or...
    * Each agent (or categories of agents)

## Rules

* Typically based around 'if-else' statements
    * Agents carry out action once specified condition is satisfied
    * Can also be put into action in ignorance of actions of other agents
* Can also be embedded with notion of *learning* and thus 'intelligence' through evolutionary computation!

## Agents as Actors

* Agents can interact with each other and amongst themselves and with the environment
* Relationships range from *reactive*
    * i.e. only perform actions when triggered by external stimulus e.g. actions of other agent
* to *goal*-directed
    * i.e. seeking a particular goal

## Actor Behavior

* Behavior can be scheduled to take place synchronously
    * i.e. every agent performs actions at each discrete time step, all change occurs simultaneously
* or asynchronously
    * i.e. agent actions are scheduled by the actions of other agents, and/or with reference to a clock

## Agent Environments

* Environments define space in which agents operate
    * Support agent interactions with environment and other agents
* Proximity may be defined by spatial distance for continuous space, adjacency for grid cells, or by connectivity in social networks
    * Agents within environment may be *spatially explicit* though maybe *static*

## Cellular Automata (CA)

* Discrete dynamic system, the behaviour of which is specified in terms of local relations
    * Space in a CA system is divided into a lattice or grid of regularly-space cells
    * The 'state' of a cell and its behavior determined by...
        * State of nearby cells at previous time step
        * By a set of local rules and by the cell itself 

## Special Type of 'Agent'

* Automata's location does not move
    * Can only change state
    * Position of cells and neighborhood relations remain fixed
* Different from normal agents
    * Fixed in location or free to 'roam' environment
    * CAs cannot have multiple attributes
        * e.g., cell can be occupied or unoccupied

## Advantages of Agent-Based Modeling

* Moving from...
    * Aggregate to the disaggregate
    * Static to the dynamic
* 3 main claimed advantages of ABM approach over traditional models:
    * captures emergent phenomena
    * provides a natural environment for the study of certain systems
    * is flexible, particularly in relation to the development of geospatial models

| Traditional modeling       | Agent-based modeling          |
| :------------------------- | :---------------------------- |
| Deterministic (one future) | Stochastic (multiple futures) |
| Allocative (top-down)      | Aggregative (bottom-up)       |
| Equation based formulas    | Adaptive agents               |
| Do not give explanations   | Explanatory power             |
| Few parameters             | Many parameters               |
| Spatially coarse           | Spatially explicit            |
| Environment given          | Environment created           |
| You react to them          | You learn from them           |

## Capturing Emergent Behavior

1. Interaction between agents is complicated, non-linear, discontinuous, or discrete
2. Heterogeneous population of agents can lead to fluctuations (can be amplified)
    * Heterogeneity also allows for agents with varying degrees of rationality
3. The topology of agent interactions is heterogeneous and complex
4. Agents exhibit complex behavior, including learning and adaptation

## System Dynamics

> Behavior and interactions of a system allows for *system dynamics* to be directly incorporated into models

## What About Time?

* Time in ABMs is still discrete, operates in 'steps'
    * i.e. time moves in 'snapshots'
* Time steps may be small $\rightarrow$ \~real time dynamics
    * Different processes occur over different time periods
    * e.g., long term economic cycles, daily commuting, hour by hour social interactions
* Incorporate *different* scales of time/processes into single simulation

## Intuitive?

* System composed of real-world entities using object-orientated principles
    * More akin to 'reality' than other modeling approaches?
* You like social theory?
    * Test social theories which cannot easily be described using math
* Useful when natural to describe constituent units of system rather than system directly
    * Activities are (arguably) more natural way to describe system

## Useful When...

* Behavior of individuals cannot clearly be defined through aggregate transition rates
    * e.g. decision to move
* Individual behavior is complex
    * *Could* do it with equation, but equations can become intractable
* Agent behavior is stochastic
    * Randomness applied *strategically*, rather than *arbitrarily* within aggregate equations

## What About Data?

* No need for data?!
    * This is maybe a limitation, but also a strength...

## Flexibility

* Benefit from mobility that ABMs offer
* Defined within *any* given system environment
    * e.g. a building, a city, a road network, a computer network, etc
    * Essentially *without* scale
* Combine objects to represent phenomena at different scales within same model
    e.g., families in houses, employees in office, all commute to city, etc

## Facilitates Exploration

* Robust and flexible framework for *tuning* complexity of agents
    * i.e. behavior, rationality, learning and evolution, and rules of interaction
* Ability to adjust levels of description and aggregation
    * Experiment with aggregate agents, sub-groups of agents, single agents
    * All with different levels of description coexisting
* ABMs useful when appropriate level of description unknown

## Limitations of Agent-Based Modeling

* Same old same old...
    * Model is only as useful as purpose for which it's constructed
        * Built at right level of abstraction for every phenomenon
        * Using right amount of detail for model to serve purpose
* Dynamics not always what we're looking for
    * Nature of system being modeled is important consideration

## Disaggregation

* By definition, ABMs consider systems at disaggregated level
    * Description of many agent attributes and behaviors, and interactions with environment    
* Issues
    * Extremely computational
    * Hard to predict aggregate outcomes
    * Limit to explorable parameter space
    * Hard to assess robustness of results

## Garbage in Garbage Out

* Individual level data is hard to come by!
    * So difficult to use data to drive individual behavior?
    * Although, maybe don't need data here?
* Too surprising sometimes?
    * Sensitive to initial conditions and small variations in interaction rules
    * Not great for simple processes that don't display complex behavior

## Highlight Uncertainty

* ABMs *do* highlight uncertainty behind modeling geographical systems
* Generate *possible* futures rather than definitive models with strong predictive assumptions
    * This is actually a good thing!
